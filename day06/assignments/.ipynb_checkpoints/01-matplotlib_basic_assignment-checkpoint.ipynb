```python
import pandas as pd
df = pd.read_csv("../data/company_sales_data.csv")
df
```

## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
from matplotlib import pyplot as plt

plt.plot(df['month_number'],df['total_profit'])
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Profit in Month')
plt.xticks(df['month_number'])
plt.ylim(100000,500000)



plt.show()

```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

<img src="../data/company_sales_data_of_last_year.png" alt="company_sales_data_of_last_year" style="width: 700px;"/>


```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8,7),linewidth= 3)
# mfc : markerfacecolor ; ms : markersize 
plt.plot(df['month_number'],df['total_profit'], linestyle = 'dotted',linewidth = 3,color = 'r', marker = '.',mfc ='black', ms = 10)
# types of writing styles are : italic , normal and oblique
plt.title('Company Sales data of last year', style = 'oblique')

plt.xlabel('Month Number',style= 'italic')
plt.ylabel('Profit in Month',style= 'italic')

plt.xticks(df['month_number'])
plt.ylim(100000,500000)
plt.legend(['Profit data of last year'],loc = 'lower right')



plt.show()
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
plt.figure(figsize=(7,5))

plt.plot(df['month_number'],df['facecream'], color = 'pink')
plt.plot(df['month_number'],df['facewash'], 'brown', label = 'facewash')
plt.plot(df['month_number'],df['toothpaste'], 'red',label = 'toothpaste')
plt.plot(df['month_number'],df['bathingsoap'], 'green', label = 'bathingsoap')
plt.plot(df['month_number'],df['shampoo'], 'black', label = 'shampoo')
plt.plot(df['month_number'],df['moisturizer'], 'blue',label = 'moisturizer')
plt.title('Company Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(df['month_number'])




plt.show()
```

## Exercise 4
**Read toothpaste sales data of each month and show it using a scatter plot**
Also, add a grid in the plot. gridline style should “–“.

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
plt.figure(figsize = (5,5))
width = 0.3


plt.bar(df['month_number']-width/2 ,df['facecream'],width , label = 'Face Cream sales data', color = 'r' )
plt.bar(df['month_number']+width/2 ,df['facewash'],width , label = 'Face Wash sales data' , color = 'blue')
plt.legend(['Face Cream sales data','Face Wash sales data'] , loc = 'upper left')

plt.title('Facewash and Faceream sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.grid(True,linewidth=0.5,color='#aaaaaa',linestyle='dashdot')

plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
def units_per_year(my_df):
    total = 0
    for i in my_df:
        total += i
    return total
units_list = [units_per_year(df["facecream"]), units_per_year(df["facewash"]), units_per_year(df["toothpaste"]), units_per_year(df["bathingsoap"]), units_per_year(df["shampoo"]), units_per_year(df["moisturizer"])]
labeles = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]
plt.figure(figsize=(10,12))
plt.pie(x=units_list, labels=labeles, autopct="%.1f%%")
plt.legend(loc="lower right")
plt.show()
```
